```python
import torch
from torch.utils.data import DataLoader
import temana as tm
```

```python
import matplotlib.pyplot as plt
```

```python
batch_size = 128
epochs = 1000
learning_rate = 0.001
criterion = tm.nn.MSELoss()

model1 = tm.CNNAutoencoder(n_input=1000, n_hidden=128, n_output=5000)
# Lectura de los coeficientes calculados
model1.load_state_dict(torch.load("CNNmodelAVFV_1000_5000.ckpt"))
x_valid = tm.read_data('SamplesAV_FV2024_07_12/SignalAVFV_Sub_Sample1000_5000Val.txt')
y_valid = tm.read_data('SamplesAV_FV2024_07_12/SignalAVFV_Super_Sample150_250_500_1000_5000Val.txt')
# Carga del dataset de validación
valid_dataset = tm.MyDataset(x_valid.unsqueeze(2), y_valid.unsqueeze(2))
valid_loader = DataLoader(valid_dataset, batch_size=batch_size, shuffle=False)

# Evaluación del modelo

with torch.no_grad():
    model1.eval()
    total_loss = 0
    for batch_idx, (data, target) in enumerate(valid_loader):
        output = model1(data)
        loss = criterion(output, target)
        total_loss += loss.item()

    print(f"Pérdida promedio de validación: {total_loss / len(valid_loader)}")
```
